```python
import pandas as pd
import os
import numpy as np


#from google.colab import drive
#drive.mount('/content/gdrive')
os.chdir('C:\insta\data')
```

```python
# feature built based on users
df_user = pd.read_csv('df_user.csv')
#df_user = pd.read_csv('gdrive/My Drive/Colab Notebooks/instacart/df_user.csv')
df_user.head()
```

```python
# features built based on products
df_products_merged = pd.read_csv('df_products_merged.csv', encoding = 'latin1')
#df_products_merged = pd.read_csv('gdrive/My Drive/Colab Notebooks/instacart/df_products_merged.csv')
df_products_merged = df_products_merged.drop(['product_name'], axis = 1)
df_products_merged.head()
```

```python
# feature built based on the specific user and product
df_user_product = pd.read_csv('df_user_product.csv')
#df_user_product = pd.read_csv('gdrive/My Drive/Colab Notebooks/instacart/df_user_product.csv')
df_user_product.head()
```

```python
# merge the user and user_product based on user_id
merge_1 = pd.merge(df_user,df_user_product,on = 'user_id',how = 'inner')
del df_user
del df_user_product
merge_1.head()
```

```python
# then merger previous data with products based on product_id
final_merge = pd.merge(merge_1, df_products_merged, on = 'product_id', how = 'inner')
del df_products_merged

final_merge.sample(10)
```

```python
final_merge.shape
```

### Creating Training 

```python
# to make train data, need to combine train data from orders.csv
orders = pd.read_csv("orders.csv")
#orders = pd.read_csv('gdrive/My Drive/Colab Notebooks/instacart/orders.csv')
orders.head()
```

```python
# combine with the last orders(it contains both train and test)
orders_last = orders.loc[((orders.eval_set == 'train') | (orders.eval_set == 'test')), ['user_id', 'eval_set', 'order_id', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']]
del orders
orders_last.head()
```

```python
#merging the orders_last with feature data (final_merger)
final = final_merge.merge(orders_last, on='user_id', how='left')
del final_merge
del orders_last
final.head()
```

```python
#df_dow = pd.read_csv("gdrive/My Drive/Colab Notebooks/instacart/df_dow.csv")
#df_hour = pd.read_csv("gdrive/My Drive/Colab Notebooks/instacart/df_hour.csv")
df_dow = pd.read_csv("df_dow.csv")
df_hour = pd.read_csv("df_hour.csv")
final = final.merge(df_dow, on = 'order_dow', how='inner')
final = final.merge(df_hour, on = 'order_hour_of_day', how='inner')
del df_dow
del df_hour

final.sample(10)
```

```python
test = final[final.eval_set == 'test']

#test.drop(['eval_set', 'order_id'], axis = 1, inplace = True)
test = test.drop(['eval_set', 'order_id'], axis = 1)

test = test.replace(np.nan, 0)

#test.to_csv('gdrive/My Drive/Colab Notebooks/instacart/test.csv', index = False)
test.to_csv('test.csv', index = False)
test.head()
```

```python
print(test.shape)
print(test.isnull().values.any())
```

```python
# train dataset
train = final[final.eval_set == 'train']
train.head()
```

```python
#merging the information contained in the order_products__train.csv into data_train.
order_products_train = pd.read_csv("order_products__train.csv")
#order_products_train = pd.read_csv('gdrive/My Drive/Colab Notebooks/instacart/order_products__train.csv')
order_products_train.head()
```

```python
#merging the information contained in the order_products__train.csv into train.
train = train.merge(order_products_train[['product_id', 'order_id', 'reordered']], on=['product_id', 'order_id'], how='left')
del order_products_train
train.head()
```

```python
#replace the NAN values with 0 in the reordered
train.reordered.fillna(0, inplace=True)
train.replace(np.nan, 0, inplace = True)
```

```python
#deleting eval_set, order_id as they are not needed for training.
#train.drop(['eval_set', 'order_id'], axis = 1, inplace = True)
train = train.drop(['eval_set', 'order_id'], axis = 1)

#train.to_csv('gdrive/My Drive/Colab Notebooks/instacart/train.csv', index = False)
train.to_csv('train.csv', index = False)

#setting user_id and product_id as index.
#train = train.set_index(['user_id', 'product_id'])

train.head()
```
